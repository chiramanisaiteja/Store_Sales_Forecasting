```python
oil.info()

oil.isnull().sum()

oil.duplicated().sum()

oil.describe()
oil.head()
```

```python
oil_cleaned =oil.copy()
```

```python
oil_cleaned['date'] = pd.to_datetime(oil_cleaned['date'])
```

```python
# Check missing values before cleaning
print("Missing values before:")
print(oil_cleaned.isnull().sum())
```

```python
# Fill missing oil prices using interpolation
oil_cleaned['dcoilwtico'] = oil_cleaned['dcoilwtico'].interpolate(method='linear')
```

```python
# Check missing values after cleaning
print("\nMissing values after:")
print(oil_cleaned.isnull().sum())
```

```python
# Verify the data types
oil_cleaned.info()
```

```python
oil_cleaned['dcoilwtico'].fillna(93.14, inplace=True)
```

```python
oil_cleaned.info()
```

```python
oil_cleaned['dcoilwtico'] = (
    oil_cleaned['dcoilwtico']
    .interpolate(method='linear')
    .bfill()
)
```

```python
oil_cleaned.info()
```

```python
oil_cleaned.to_csv("oil_cleaned.csv", index=False)
```
