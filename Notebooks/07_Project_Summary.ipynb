# 1. Business Problem

Retail companies need accurate sales forecasts to optimize inventory, reduce stock shortages, minimize overstocking, and improve operational planning.

The objective of this project is to build a machine learning model capable of forecasting future daily sales using historical store sales data, promotions, holidays, transactions, oil prices, and store information.

The final solution follows a complete end-to-end data science workflow from raw data to model evaluation.

# 2. Dataset Description

The project combines six datasets.

| Dataset | Description |
|----------|-------------|
| train.csv | Historical daily sales |
| test.csv | Future prediction records |
| stores.csv | Store information |
| transactions.csv | Daily customer transactions |
| oil.csv | Daily oil prices |
| holidays_events.csv | Holiday and event information |

These datasets were cleaned and merged to create a single master dataset for analysis and forecasting.

# 4. Data Cleaning

The following preprocessing steps were performed:

- Converted all date columns to datetime format
- Checked for duplicate records
- Verified missing values
- Filled missing oil prices using forward fill
- Verified data types
- Prepared clean datasets for merging

Clean datasets were saved for the next stage of the project.


# 5. Data Merging

The cleaned datasets were merged into one master dataset.

Merge Keys:
- date
- store_nbr

Datasets merged:
- Train
- Stores
- Transactions
- Oil Prices
- Holiday Information

The final master dataset contains sales, promotions, store details, transaction activity, oil prices, and holiday-related features.

# 6. Exploratory Data Analysis (EDA)

The following analyses were performed:

- Daily sales trend
- Monthly sales trend
- Top-performing stores
- Top-selling product families
- Promotion impact on sales
- Holiday vs normal day sales
- Day-of-week sales analysis
- Transactions vs sales relationship

### Key Findings

- Sales generally increased over time.
- Grocery I generated the highest sales.
- Promotions significantly increased sales.
- Holiday sales were slightly higher than normal days.
- Sunday recorded the highest average sales.
- Higher transaction volume generally resulted in higher sales.

# 8. Model Training

Two machine learning algorithms were evaluated.

| Model | MAE | RMSE | R² |
|------|------:|------:|------:|
| Random Forest | 58.21 | 213.54 | 0.9742 |
| XGBoost | 57.42 | 216.48 | 0.9735 |
| XGBoost Light | 59.35 | 217.75 | 0.9732 |

A time-based train-test split was used to ensure that future data was never used during training.

# 9. Final Model

The final implementation uses the XGBoost Light model.

Training Records:
- 150,000

Testing Records:
- 50,000

Final Performance

- MAE = 59.35
- RMSE = 217.75
- R² = 0.9732

The model successfully explained approximately 97.3% of the variation in daily sales while remaining computationally efficient within Google Colab memory limitations.

# 10. Model Evaluation

Model performance was evaluated using:

- RMSE (Root Mean Squared Error)
- R² Score

Evaluation visualizations include:

- Actual vs Predicted Sales

- Model Error Metrics

Most prediction errors were concentrated close to zero, indicating good forecasting accuracy.
